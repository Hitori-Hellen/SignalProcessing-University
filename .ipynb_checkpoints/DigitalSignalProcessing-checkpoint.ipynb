**HW 1**

```python
import numpy as np
import seaborn as sns
import soundfile as sf
import matplotlib.pyplot as plt
from IPython.display import Audio
import pandas as pd
```

```python
mean = 0.5
std = 1
n = 100000
wave = np.random.normal(mean, std, size=n)

plt.plot(wave)
plt.show()

sf.write('audio_noise.wav', wave, 44100, subtype='PCM_24')

Audio('audio_noise.wav')
```

```python
plt.plot(np.random.normal(mean, std, size=n))
plt.show()
```

```python
sns.set_theme()

sns.relplot(
    data=wave,
    kind='line',
)
```

**HW 2**

```python
data, fs = sf.read('handel.wav', dtype='float32')
data = np.array(data)

sns.set_theme()

sns.relplot(
    data=data,
)

plt.show()
```

```python
data
```

```python
Audio(data=data, rate=fs)
```

```python
len_data = int(len(data) / 2)
len_data
sample = []
for i in range(0, len_data):
    sample.append(data[(2*i)-2])
    
sample = np.array(sample)

sns.set_theme()

sns.relplot(data=sample)

plt.show()
```

```python
Audio(data=sample, rate=fs/2)
```

```python
len_data = int(len(data) / 4)
len_data
sample_c = []
for i in range(0, len_data):
    sample_c.append(data[(4*i)-4])
    
sample_c = np.array(sample_c)

sns.set_theme()

sns.relplot(data=sample_c)

plt.show()
```

```python
Audio(data=sample_c, rate=fs/4)
```
